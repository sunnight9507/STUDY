### 추천시스템 Basic Review

> #### 1) 추천 시스템 등장
> - 인터넷은 폭발적인 성장과 수많은 정보가 쏟아지는 공간
> - 인터넷 관련 비즈니스 활성화
    - e-Business, e-Commerce 등
> - 수많은 정보(+데이터)로 사용자가 적절한 결정을 내리기 어려움
    
> ####  2) 추천시스템 정의
> - 사용자(User)로부터 선호도(preferences)여부를 데이터화
    - Explicit Feedback(평점 등)
    - Implicit Feedback(사용자의 행동 패턴 등)
> - 사용자의 선호도와 제한사항(constraints)를 바탕으로 가장 적절한 아이템 점수 계산
> - 사용자에게 가장 관련 있는 아이템 순서대로 리스트 형태로 제시
    
> #### 3) 추천시스템 연구 현황
> - 정보검색 분야와 비슷하지만, 비교적 새로운 연구 분야
> - 최근 다양한 형태의 추천시스템 연구 진행중
> - 여러 기업과 학회에서 다양하게 연구 진행중
    - SIGAR, WWW, ACM RecSys, IEEE 등
    - 유튜브, 넷플릭스, 왓챠 등
> - Netflix Prize 추천대회
    
### 추천시스템을 왜 사용해야할까?

> - 더 많은 아이템을 판매할 수 있다.
> - 더 다양한 아이템을 판매할 수 있다.
> - 소비자 만족도가 증가한다.
> - 충성도 높은 고객이 증가한다.
> - 고객이 원하는 것이 무엇인지 알 수 있다.

### 추천시스템 예시

> - 넷플릭스 : 콘텐츠 추천
> - 인스타그램 : 팔로우 추천
> - 쿠팡 : 상품 추천

### 추천시스템 : 데이터 종류

> #### 1) Items
> - 추천할 항목
> - 가치와 데이터의 복잡함에 따라 구분 가능
    - Low : 뉴스, 책, 영화 등
    - High : 금융 상품, 직업, 여행 등
        
> #### 2) User
> - 여러 목적과 특징을 갖는 집단
> - 행동 패턴 등으로 유저 모델링
> - 협업 필터링 : 여러 아이템에 대한 평점 리스트
> - 사회적인 요인을 활용한 추천시스템
    
> #### 3) Transactions
> - 유저와 아이템간의 상호작용
> - 유저가 인터넷에 남긴 로그에서 중요한 정보 추출
> - 해시태그, 평점, Implicit한 정보 모두 활용 가능

### 추천시스템이 풀고자 하는 문제

> #### 1. 랭킹 문제
> - 특정 유저가 특정 아이템에 대한 평점(or 점수)를 정확하게 예측할 필요 없음
> - 특정 유저, 아이템을 좋아할만한 Top-K 유저, 아이템 선정

> #### 2. 예측 문제
> - 유저-아이템 조합에서 평점(or 점수)를 예측
> - User-Item(mxn) 행렬을 채우는 문제
    - 관측값(Observed value)은 모델 학습에 사용
    - 결측값(Missing value)는 모델 평가에 사용
