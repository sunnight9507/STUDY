추천시스템의 기본 모델은 **등급 또는 구매 행동과 같은 사용자와 item간의 관계**와 **사용자 및 item에 대한 속성 정보(ex: 기본 정보, 관련 키워드)** 라는 두 가지 종류의 데이터로 작동된다. 전자를 사용하는 방법을 **`협업 필터링`**이라고 하는 반면 후자를 사용하는 방법은 **`콘텐츠 기반 추천시스템`** 이라고 한다. 콘텐츠 기반 시스템도 대부분의 경우 rating 매트릭스를 사용하지만 일반적으로 모든 사용자의 등급이 아닌 단일 사용자의 등급에 중점을 둔다. 지식 기반 추천시스템에서 추천은 명시적으로 지정된 사용자 요구에 기초한다. 과거 rating을 사용하거나 데이터를 구입하는 대신 외부 지식 기반과 제약 조건을 활용해 추천을 한다. **일부 추천 시스템은 이러한 다양한 측면을 결합해 하이브리드 시스템을 만든다.** 하이브리드 시스템은 다양한 유형의 추천시스템의 강점을 결합하여 다양한 설정에서 보다 견고하게 수행할 수 있는 기법을 만들 수 있다. 다음에서는 이러한 기본 모델에 대해 간략하게 논의하고, 또한 이 모델에 대해 논의되는 책 속의 관련 장에 대한 포인터를 제공할 것이다.

# 1.3.1 Collaborative Filtering Models

> **협업 필터링 모델**은 **여러 사용자가 제공 한 평가의 협업 기능을 사용하여 추천**

> 설계할 때 가장 큰 어려움은 **rating 매트릭스가 희박하다는 것**
- ex) 사용자가 특정 영화에 대한 호불호를 나타내는 rating을 지정하는 영화 앱
- 대부분의 사용자들은 이용할 수 있는 영화의 넓은 영역의 극히 일부분만 보았을 것이다.
- 결과적으로 대부분의 rating이 지정되지 않는다.
    
> 기본 아이디어는 **관찰된 등급이 다양한 사용자 및 항목에 걸쳐 높은 상관 관계를 보이는 경우가 많기 때문에 관찰되지 않은 rating을 대체 할 수 있다는 것**이다.
- ex) 예를 들어 취향이 매우 비슷한 Alice와 Bob이라는 두 명의 사용자를 생각해보자.
- 둘 다 관찰된 등급이 매우 유사한 경우 기본 알고리즘으로 유사성을 확인 할 수 있다.
- 이러한 경우, 그들 중 **하나만 값을 지정한 등급도 유사할 가능성이 매우 높다.**
- 이 유사성은 불완전하게 지정된 값에 대해 추론하는 데 사용될 수 있다.


> - 협업 필터링을 위한 대부분의 모델은 **예측 프로세스를 위해 항목 간 상관 관계 또는 사용자 간 상관 관계를 활용**하는 데 중점을 둔다.
- 일부 모델에서는 두 가지 유형의 상관 관계를 모두 사용한다. 
- 또한 일부 모델은 분류 기가 레이블이 지정된 데이터에서 학습 모델을 만드는 것과 거의 동일한 방식으로 신중하게 설계된 최적화 기술을 사용하여 학습 모델을 만듭니다.
- 그런 다음 이 모델은 분류기가 누락 된 테스트 레이블을 입력하는 것과 동일한 방식으로 행렬에서 누락 된 값을 채우는 데 사용된다.

일반적으로 사용되는 방법에는 **`메모리 기반 방법`**과 **`모델 기반 방법`** 두 가지가 있다.

### 1) Memory-based methods

> - 메모리 기반 방법은 이웃 기반 협업 필터링(neighborhood-based collaborative filtering) 알고리즘이라고도 한다.
- 사용자 항목 조합의 rating이 이웃을 기반으로 예측되는 최초의 협업 필터링 알고리즘 중 하나이다.
- 두 가지 방법으로 정의되어 있다.

> ### User-based collaborative filtering
- 이 경우 사용자 A와 비슷한 생각을 가진 사용자의 rating을 이용하여 A에게 추천을 한다.
- **기본 아이디어는 사용자 A와 유사한 사용자들을 결정하고 이 그룹의 rating에 대한 가중 평균을 계산해 A의 관찰되지 않은 rating을 예측하는  것이다.**
    - 예를 들어 Alice와 Bob이 과거에 비슷한 방식으로 영화를 평가했다면 영화 터미네이터에서 Alice의 관찰 된 등급을 사용하여 이 영화에 대한 Bob의 관찰되지 않은 등급을 예측할 수 있다.
    - 일반적으로 Bob과 **가장 유사한 k명의 사용자를 사용**하여 Bob에 대한 평가를 예측할 수 있습니다.
    

> ### Item-based collaborative filtering
- 사용자 A에 의한 대상 항목 B에 대한 등급 예측을 하기 위해서는, 대상 항목 B와 가장 유사한 항목의 집합 S를 결정하는 것이 첫 번째 단계이다.
- 집합 S에 대한 A의 rating은 사용자 A가 B항목을 좋아할지 여부를 예측하는 데 사용된다.
    - 예를 들어 Alien 및 Predator와 같은 유사한 공상 과학 영화에 대한 Bob의 rating을 사용하여 Terminator에 대한 평가를 예측할 수 있다.
    
> - **`장점`**은 **구현이 간단**하고 **추천의 결과에 대해 설명하기 쉽다는 것**이다.
- **`단점`**은 **sparse한 매트릭스에서 잘 작동하지 않는다.**


### 2) Model-based methods

> - 모델 기반 방법에서 기계 학습 및 데이터 마이닝 방법은 예측 모델의 맥락에서 사용된다.
- 모델이 매개변수화된 경우, 이 모델의 매개변수는 최적화 프레임워크의 맥락 안에서 학습된다.
- 모델 기반 방법의 일부 예에는 의사결정 나무, 규칙 기반 모델, 베이지안 방법 및 잠재 요인 모델이 포함된다.
- 잠재 요인 모델과 같은 이러한 방법 중 대부분은 **희소 등급 행렬에 대해서도 높은 수준의 적용 범위를 갖는다.**

## 1.3.1.1 Types of Ratings

- 추천 알고리즘의 설계는 평가 추적에 사용되는 시스템의 영향을 받는다. 
- rating은 종종 해당 항목의 호불호 수준을 나타내는 척도로 지정된다.
- 가능한 각 등급에 따라 **사용자의 관심 수준에 대한 의미적 해석을 표시한 것**이다.
- 일반적으로 **rating은 `구간`을 기준**으로 하며, 여기서 **순서의 이산형 집합을 사용하여 호불호를 정량화**한다.
    - Jester joke recommendation engine과 같이 -10~10 사이의 연속 값일 수도 있다. 하지만 이와 같은 경우는 드물다. 
- 가능한 등급의 수는 현재 시스템에 따라 상이하다.
    - 경우에 따라서는 평점이 짝수일 수도 있고, 중립 등급이 누락될 수도 있다. 이 접근방식을 강제선택등급제라고 한다.
    
> #### Examples of Explicit and Implicit Ratings
- 일반적으로 등급은 m×n 매트릭스로 나타낼 수 있으며, 여기서 m과 n은 일반적으로 매우 크고 수십만 개의 순서로 범위가 될 수 있다.
- 등급이 단항인 경우 매트릭스는 긍정적 선호도만 허용된다. 보통 단일 등급 매트릭스라 한다.
    - 일반적으로 **단일 행렬은 아이템 구매와 같은 사용자 행동에 의해 생성**되며, **implicit feedback matrices**라고도 한다.
    - 단항 rating은 사용자가 항목을 싫어하는지 여부에 대한 정보를 사용할 수 없기 때문에 현재 추천 알고리즘에 상당한 영향을 미친다.
    - 단항 행렬의 경우 초기 단계에서 누락된 항목을 0으로 처리하여 간단한 방법으로 분석을 수행하는 것이 좋다.
    - 실제로 누락된 항목이 0으로 대체되지 않을 경우 상당한 오버피팅이 가능하다.
- Explicit ratings 매트릭스에서는 누락된 등급의 사전 개정이 권장되지 않는다.
    - 호불호가 모두 있는 명시적 등급 매트릭스에서 누락된 항목을 임의의 값(예: 0 또는 행/열/데이터 평균)으로 대체하면 항상 분석에서 상당한 양의 치우침이 발생한다.
    
## 1.3.1.2 Relationship with Missing Value Analysis

- 협업 필터링 모델은 결측값 분석과 밀접한 관련이 있다.
- 협업 필터링은 기본 데이터 행렬이 매우 크고 희박한 이 문제의 특별한 경우로 볼 수 있다.
- 실제로 최근 잠재 요인 모델과 같은 추천 시스템 모델의 일부 클래스는 결측치 분석의 맥락에서 일찍 연구되었다.

## 1.3.1.3 Collaborative Filtering as a Generalization of Classification and Regression Modeling(분류 및 회귀모형의 일반화로서 협업 필터링)

- 협업 필터링 방법은 분류 및 회귀 모델링의 일반화로 볼 수 있다.
- 분류 및 회귀 모델링 문제에서 클래스 / 종속 변수는 결측값을 가진 속성으로 볼 수 있다.
- 다른 열은 피쳐/독립 변수로 처리된다.
- 추천 문제에서는 각 특징이 종속 변수와 독립 변수의 이중 역할을 하기 때문에 클래스 변수와 특징 변수 사이에 명확한 구별이 존재하지 않습니다.
- 또한, 작업 수행 시 수행 중인 훈련 행과 테스트 행의 구분이 없습니다.왜냐하면 행에 누락된 항목이 포함될 수 있기 때문이다.
- 협업 필터링은 예측이 행 방식보다는 엔트리 방식으로 수행되는 분류 / 회귀 모델링의 일반화이다.
- 분류/회귀 모델링과 협력 필터링의 관계는 분류 및 회귀 모델링 방법의 많은 원리가 추천 시스템으로 일반화될 수 있기 때문에 명심해야 한다.
- 행렬을 채우는 문제는 분류 및 회귀에서 transductive 설정과 많은 특성을 공유한다.
- **Transductive Learning vs Inductive Learning**

# 1.3.2 Content-Based Recommender Systems

> **콘텐츠 기반 추천 시스템에서는 `항목별 설명`을 사용해 추천**
- 사용자의 등급과 구매 행태가 아이템에서 이용 가능한 콘텐츠 정보와 결합된다.
- ex) 존이 영화 터미네이터를 높게 평가했지만 우리는 다른 사용자의 등급에 접근할 수 없는 상황
- 협업 필터링 방법은 배제된다. 그러나 터미네이터라는 영화 속성에는 에일리언, 프레데터 등 다른 공상과학 영화와 비슷한 장르 키워드가 담겨 있다. 이런 경우에, 이러한 장르 키워드를 가진 영화들을 존에게 추천할 수 있다.

> 콘텐츠 기반 방법에서 등급으로 레이블이 지정된 항목 설명은 사용자별 분류 또는 회귀 모델링 문제를 생성하기 위한 훈련 데이터로 사용된다.각 사용자에 대해 훈련 데이터(문서)는 자신이 구입했거나 등급을 매긴 항목에 대한 설명에 해당된다.
클래스(또는 종속변수)는 지정된 등급 또는 구매 행동에 해당한다.
이 사용자별 모델은 해당 개인이 자신의 등급이나 구매 행동을 알 수 없는 항목을 좋아할지를 예측하는 데 사용된다.

> 콘텐츠 기반 방법은 해당 항목에 대한 충분한 rating 데이터를 사용할 수 없는 경우 새로운 항목에 대한 추천을 진행할 때 몇 가지 장점이 있다. 이는 유사한 속성을 가진 다른 항목이 활성 사용자에 의해 등급이 매겨졌을 수 있기 때문이다.

> 컨텐츠 기반 방법에는 다음과 같은 몇 가지 단점도 있다.
1. 대부분의 경우 콘텐츠 기반 방법은 키워드 또는 콘텐츠 사용으로 인해 **확실한 추천**을 한다. 예를 들어 **사용자가 특정 키워드 집합의 항목을 소비한 적이 없는 경우 그러한 항목은 추천될 가능성이 없다.** 구축된 모델은 현재 이용자에게만 한정되어 있으며, 유사한 이용자의 커뮤니티 지식은 활용되지 않기 때문이다. 이 현상은 **추천 품목의 다양성을 줄이는 경향이 있다.**
2. 컨텐츠 기반 방법은 새로운 항목에 대한 권장사항을 제공하는 데 효과적이지만, **새로운 사용자에게 권장사항을 제공하는 데는 효과적이지 않다.** 이는 대상 사용자의 학습 모델이 자신의 평가 기록을 사용해야 하기 때문이다. 사실, 과적합없는 정확한 예측을 하기 위해서는 일반적으로 대상 사용자가 사용할 수 있는 많은 등급을 갖는 것이 중요하다.

따라서 콘텐츠 기반 방법은 협업 필터링 시스템과 다른 장단점이 있다.

> 앞서 언급한 설명은 콘텐츠 기반 방법에 대한 기존의 학습 기반 관점을 제공하지만, 이러한 방법에 대한 더 넓은 관점이 때때로 사용된다. 예를 들어, **사용자는 자신의 프로필에 관련 키워드를 지정**할 수 있다. 이러한 프로필 키워드는 추천을 하기 위해 항목 설명과 일치하게 만들수 있다. 이러한 접근 방식은 추천 프로세스에서 등급을 사용하지 않으므로 **콜드 스타트에서 유용**하다. 그러나 유사성 메트릭은 종종 도메인 지식에 기반하기 때문에 이러한 방법은 종종 **지식 기반 시스템으로 알려진 추천 시스템의 별개의 클래스로 간주**됩니다. 지식 기반 추천시스템은 콘텐츠 기반 추천시스템과 밀접한 관련이 있는 것으로 간주되는 경우가 많으며, 두 가지 방법 사이에 명확한 구분이 존재하는지 의문을 제기하기도 한다.

# 1.3.3 Knowledge-Based Recommender Systems

> - **지식 기반 추천자 시스템은 특히 자주 구매하지 않는 항목에서 유용**하다.
    - 예를 들어 부동산, 자동차, 관광 요청, 금융 서비스 또는 고가 사치품 등의 품목이 포함된다.
    - 이러한 경우 추천 프로세스에 대해 충분한 rating이 제공되지 않을 수 있다.
    - 아이템을 거의 구입하지 않고, 상세한 옵션의 종류가 다르기 때문에, 당면한 아이템의 특정 인스턴스화(즉, 옵션의 조합)에 대한 충분한 등급을 얻기 어렵다.
    - 이 문제는 추천 프로세스에 대해 충분한 등급을 사용할 수 없는 경우 콜드 스타트 문제가 발생한다.
- 또한, **소비자 선호의 본질**은 그러한 항목을 처리할 때 **시간이 지남에 따라 진화**할 수 있다.
    - 예를 들어, 자동차의 모델은 몇 년 동안 크게 진화할 수 있으며, 그 결과 선호도도 그에 상응하는 진화를 보여줄 수 있다.
- 특정 아이템은 그것의 **다양한 속성에 대응하는 속성을 가질 수 있고**, 사용자는 **특정한 속성을 가진 아이템에만 관심을 가질 수 있다.**
    - 예를 들어, 자동차는 모델, 색상, 엔진 옵션 및 실내 옵션 등 여러 가지 제품을 가질 수 있으며, 사용자의 관심사는 이러한 옵션의 매우 구체적인 조합에 의해 규제될 수 있다.
    - 이러한 경우 항목 도메인은 다양한 속성 측면에서 복잡해지는 경향이 있으며, 적절한 등급을 많은 수의 조합과 연관시키기가 어렵다.
- => 이러한 경우에 추천 목적으로 **rating을 사용하지 않는 `지식 기반 추천 시스템`**으로 해결할 수 있습니다.

> - **`지식 기반 추천 프로세스`**는 **고객 요구 사항과 항목 설명 사이의 유사성** 또는 **사용자 요구 사항을 지정하는 제약 조건의 사용에 기초하여 수행**된다.
- 이 프로세스는 사용할 규칙 및 유사성 기능에 대한 데이터를 포함하는 지식 기반을 사용하여 진행된다.
- 요구 사항을 명시적으로 지정하면 추천 프로세스에 대한 사용자 제어가 향상됩니다.
- 지식 기반 시스템은 사용자가 원하는 것을 명시적으로 지정할 수 있다는 점에서 독특하다.

- 지식 기반 추천 시스템은 사용되는 인터페이스 유형(및 해당 지식)에 따라 분류할 수 있다.

> ### 1) Constraint-based recommender systems
- **`제약조건 기반` 시스템**에서 사용자는 **일반적으로 항목 속성에 대한 요구사항이나 제약조건(예: 하한 또는 상한)을 지정**한다.
- 도메인별 규칙은 사용자 요구 사항을 항목 속성과 일치시키기 위해 사용된다.
- 이러한 규칙은 시스템에서 사용하는 도메인별 지식을 나타내며 항목 속성에 대한 도메인별 제약조건의 형태를 취할 수 있다.
    - ex) 1970년 이전의 자동차에는 크루즈 컨트롤이 없다.
- 더욱이 제약에 기반한 시스템은 종종 사용자 속성과 항목 속성에 관련된 규칙을 만든다.
    - ex) 투자를 오래한 사람은 위험이 있는 상품에 투자하지 않는다.
- 검색 프로세스에서 사용자 속성을 지정할 수 있다. 
    - 반환된 결과의 수와 유형에 따라 사용자는 원래 요구 사항을 수정할 기회가 있을 수 있다.
    - 예를 들어, 너무 적은 수의 결과가 반환될 때 일부 제약조건을 완화하거나 더 많은 제약조건을 추가할 수 있다.
    - 이 검색 프로세스는 사용자가 원하는 결과를 얻을 때까지 인터렉리브하게 반복된다.
    
> ### 2) Case-based recommender systems
- **`사례기반` 추천시스템**의 경우, **사용자가 특정 사례를 대상 또는 기준점으로 지정**한다.
- 유사성 메트릭은 도메인별 방식으로 신중하게 정의되는 경우가 많다.
- 반환된 결과는 사용자가 일부 대화형 수정을 통해 새로운 대상 사례로 사용되는 경우가 많다.
    - 예를 들어 사용자가 원하는 결과와 거의 유사한 반환된 결과를 볼 때 해당 대상으로 쿼리를 다시 실행할 수 있지만 일부 속성은 사용자가 원하는대로 변경된다.
    - 이 대화형 프로세스는 사용자가 관심 있는 항목으로 안내하는 데 사용된다.

> - 두 경우 모두 **사용자에게 지정된 요구사항을 변경할 수 있는 기회를 제공**한다.
- 그러나 이것이 수행되는 방식은 두 경우에서 다르다.
- **`사례기반 시스템`**에서는 유사도 메트릭과 함께 검색을 안내하기 위해 **예제(또는 사례)를 기준점으로 사용**한다.
- 사용자가 각 반복에서 선호 항목의 하나 이상의 속성을 반복적으로 수정하는 그러한 시스템에서 피드백을 표현하는 데 인기가 있다.
- **`제약조건 기반 시스템`**에서는 검색을 안내하는 데 **규칙(또는 제약 조건)이 사용**된다.
- 안내의 형태는 종종 검색 기반 시스템의 형태를 취할 수 있으며, 여기서 사용자는 검색 기반 인터페이스로 제약 조건을 지정한다.

- 지식기반 추천시스템에서의 상호작용은 어떻게 이루어지는가? 다음 방법 중 하나 이상을 통해 수행된다.

> - #### 1) Conversational systems
- 이 경우 사용자 선호도는 **인터렉티브하게 피드백 반복을 통해 결정**된다.
> - #### 2) Search-based systems
- 검색 기반 시스템에서 사용자 선호도는 다음과 같은 **사전 설정된 질문을 사용하여 결정**된다.
    - ex) "교외 지역, 도시 안에 있는 집 어느 집을 선호합니까?"
- 경우에 따라 사용자 제약조건을 지정할 수 있는 기능을 제공하기 위해 특정 검색 인터페이스를 설정할 수 있다.
> - #### 3) Navigation-based recommendation
- 탐색 기반 권장 사항에서 사용자는 **현재 권장되는 항목에 대한 여러 변경 요청을 지정**한다.
- 반복적인 변경 요청 세트를 통해 원하는 항목에 도달 할 수 있다.
- 특정 주택을 추천할 때 사용자가 지정한 변경 요청의 예는 다음과 같다.
    - ex) "현재 추천된 집에서 서쪽으로 5마일 정도 떨어진 곳에 비슷한 집이 있었으면 좋겠다."
- 이러한 추천시스템을 'critiquing recommender systems'라고도 한다.

> - 주목할 점은 **지식기반 시스템**과 **콘텐츠 기반 시스템** 모두 **항목 속성에 크게 의존한다**는 점이다.
- 콘텐츠 특성의 사용 때문에 지식 기반 시스템은 콘텐츠 기반 시스템과 동일한 일부 단점을 상속한다.
    - 예를 들어, 콘텐츠 기반 시스템과 마찬가지로, rating의 사용이 활용되지 않기 때문에 지식 기반 시스템의 권고사항이 명확해질 수 있다.
- 지식 기반 시스템은 때때로 콘텐츠 기반 시스템의 '친척'으로 간주된다.
- 주요 **`차이점`**은 **`콘텐츠 기반 시스템`은 과거의 사용자 행동에서 배우는** 반면 **`지식 기반 추천 시스템`은 그들의 필요와 관심에 대한 능동적인 사용자 사양에 기초하여 추천**한다.

## 1.3.3.1 Utility-Based Recommender Systems

> - 유틸리티 기반 추천 시스템에서는 **사용자가 항목을 좋아할 확률을 계산하기 위해 제품 기능에 유틸리티 함수를 정의**한다.
- 유틸리티 기반 방법의 핵심 과제는 사용자에게 적절한 유틸리티 기능을 정의하는 것이다.
- 협업 필터링, 콘텐츠 기반, 지식 기반 등 **모든 추천스시템 체계**는 대상 **사용자에 대한 인식된 가치(또는 유용성)를 바탕으로 추천된 항목을 암묵적으로 순위를 매긴다.**
- 유틸리티 기반 시스템에서 유틸리티 값은 사전에 알려진 함수를 기반으로 한다.
- 이런 의미에서 그러한 기능은 일종의 외부 지식으로 볼 수 있다.
- 따라서 유틸리티 기반 시스템은 지식 기반 추천 시스템의 특정 사례로 볼 수 있다.

# 1.3.4 Demographic Recommender Systems

> - 인구 통계학적 추천 시스템의 경우, 사용자에 대한 인구 통계학적 정보는 특정 인구 통계를 등급 또는 구매 성향에 매핑할 수 있는 분류자를 학습하기 위해 활용된다.
- Grundy라고 불리는 초기 추천자 시스템은 수작업으로 조립된 고정관념의 라이브러리에 기초하여 책을 추천했다.
- 사용자의 특성은 인터렉티브한 상호작용으로 수집되었다.
- 이 연구는 마케팅 연구의 인구통계학적 집단이 아이템을 추천하는 데 사용될 수 있다는 것을 관찰했다.
- 또 다른 작업은 특정 페이지를 높게 평가한 사용자의 인구통계학적 특성에 기초하여 웹 페이지 권장사항을 작성한다.
- 많은 경우 **인구 통계 학적 정보는 추천 프로세스를 안내하기 위해 추가 컨텍스트와 결합 될 수 있다.**
- 이 접근방식은 **context-sensitive 추천시스템의 방법론과 관련**이 있다.

> - **최근의 기술은 추천을 위해 분류기를 사용하는 데 초점을 맞추고 있다.**
- 이 점에서 흥미로운 시스템 중 하나는 특정 레스토랑을 좋아할 가능성을 예측하기 위해 사용자 홈 페이지에서 기능을 추출하는 기술이었다.
- 규칙 기반 분류자는 종종 인구통계학적 프로필을 상호 작용적인 방식으로 구매 행동과 연관시키기 위해 사용된다.
- 해당 접근법은 특정 항목을 추천하는 데 특별히 사용되지 않았지만, 추천시스템과 쉽게 결합할 수 있다.
- 이러한 추천 시스템은 특징 변수가 인구 통계학적 프로파일에 대응하고 종속 변수가 등급 또는 구매 행동에 대응하는 분류 및 회귀 모델링 문제와 크게 다르지 않다.
- **인구통계학적 추천자 시스템은 일반적으로 독립적으로 최상의 결과를 제공하지 않지만, 하이브리드 또는 앙상블 모델의 구성 요소로서 다른 추천 시스템의 성능을 크게 향상시킨다.**
- 인구통계학적 기술은 때때로 지식 기반 추천시스템과 결합되어 견고성을 높인다.

# 1.3.5 Hybrid and Ensemble-Based Recommender Systems

> - 앞서 언급한 세 가지 시스템은 서로 다른 입력 소스를 이용하며, 다른 시나리오에서 잘 작동할 수 있다.
- 예를 들어, 협력 필터링 시스템은 커뮤니티 등급에 의존하고, 콘텐츠 기반 방법은 텍스트 설명과 대상 사용자 자신의 등급에 의존하며, 지식 기반 시스템은 지식 기반 맥락에서 사용자와의 상호 작용에 의존한다.
- 마찬가지로, 인구통계학적 시스템은 사용자의 인구통계학적 프로필을 바탕으로 추천을 진행한다.
- 이러한 서로 다른 시스템은 서로 다른 유형의 입력을 사용하고 서로 다른 강점과 약점을 가지고 있다.
- 지식 기반 시스템과 같은 일부 추천시스템은 상당한 양의 데이터를 사용할 수 없는 콜드 스타트 설정에서 효과적이다.
- 협업 방법과 같은 추천시스템은 많은 데이터를 사용할 수 있을 때 더 효과적이다.

> - 다양한 입력을 사용할 수있는 많은 경우 **동일한 작업에 대해 서로 다른 유형의 추천시스템을 사용할 수 있는 유연성**이 있다.
- **하이브리드 추천시스템**은 여러 유형의 머신러닝 알고리즘의 파워가 결합되어 보다 견고한 모델을 만드는 **앙상블 분석과 밀접한 관련이 있다.**
- 앙상블 기반 추천자 시스템은 여러 데이터 소스의 힘을 결합할 수 있을 뿐만 아니라, 같은 유형의 여러 모델을 결합하여 특정 클래스의 추천시스템의 효율성을 향상시킬 수 있다.
- 이 방법은 데이터 분석 분류에서 앙상블 분석과 크게 다르지 않다.

# 1.3.6 Evaluation of Recommender Systems

> - 추천 알고리즘이 얼마나 잘 수행되는지 상대적인 효과를 어떻게 평가할 수 있는지 확인이 필요하다.
- 추천시스템은 분류 및 회귀 모델링 문제와 몇 가지 개념적 유사성을 공유한다.
    - 분류 및 회귀 모델링에서 누락된 클래스 라벨은 feature를 통해 예측해야 한다.
    - 추천시스템에서 매트릭스 항목 중 하나가 누락될 수 있으므로 나머지 매트릭스의 관측된 항목에서 데이터 중심 방식으로 예측해야 한다.
    - 이런 의미에서 추천 문제는 분류 문제의 일반화로 볼 수 있습니다.
    - 따라서 **분류 모델 평가에 사용되는 많은 모델은 약간의 수정이 있더라도 추천 시스템을 평가하는 데 사용할 수 있다.**
- 평가 예측 또는 순위와 같은 추천 시스템의 다른 측면에 사용되는 평가 기술에는 상당한 차이가 있다.
    - 전자는 분류 및 회귀 모델링과 밀접하게 관련되어 있는 반면 후자는 검색 및 정보 검색 애플리케이션에서 검색 효과의 평가와 밀접하게 관련되어 있다.
