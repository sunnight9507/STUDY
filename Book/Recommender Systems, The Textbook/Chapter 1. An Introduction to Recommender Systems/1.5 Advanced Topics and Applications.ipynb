이 책은 또한 많은 주제와 응용 프로그램들을 소개할 것이다. 이 장에서는 간략하게 소개한다.

# 1.5.1 The Cold-Start Problem in Recommender Systems

> - 추천시스템의 주요 문제점 중 하나는 **초기 이용 가능한 등급의 수가 상대적으로 적다는 것**이다.
    - 이 경우 기존의 협업 필터링 모델을 적용하기가 더욱 어려워진다.
    - 콘텐츠 기반 및 지식 기반 방법이 콜드 스타트 상황에서 협업 모델보다 더 강력하지만, 그러한 콘텐츠 기반 및 지식 기반 방법이 항상 이용 가능한 것은 아닐 수 있다.
- 따라서 추천시스템의 콜드 스타트 문제를 개선하기 위해 여러 가지 구체적인 방법이 설계되었다.
- 콜드 스타트 문제에 대한 다양한 모델의 민감도는 가능한 솔루션과 함께 이 책 전체에서 강조된다.

# 1.5.2 Attack-Resistant Recommender Systems

> - **추천시스템의 사용은 다양한 제품과 서비스의 판매에 상당한 영향을 미친다.**
    - 결과적으로, 상품과 서비스의 판매자들은 추천시스템의 출력을 **`조작`**할 수 있는 상당한 경제적 동기가 있을 수 있다.
    - 그러한 조작의 한 가지 예는 추천 시스템에 자체 제품의 부풀려진 등급을 제출하는 것이다.
    - 악의적인 경쟁자는 경쟁자의 제품에 대해 편향적이고 부정적인 리뷰를 제출할 수 있다.

> - 수년에 걸쳐 **추천시스템을 공격하기 위해 수많은 정교한 전략이 개발**되었다.
- 이러한 공격은 추천 시스템의 전반적인 효율성을 감소시키고 합법적인 사용자의 경험 품질을 저하시키기 때문에 매우 바람직하지 않다.
- 따라서 이러한 공격이 있는 경우 강력한 추천을 할 수 있는 방법이 필요하다.
- 공격 방법, 공격에 대한 다양한 유형의 알고리즘의 취약성의 자세한 내용은 12장에서 다룬다.
- 또한 그러한 공격이 존재하는 곳에서 강력한 추천시스템을 구축하기 위한 다양한 전략을 제공할 것이다.

# 1.5.3 Group Recommender Systems

> - 전통적인 추천시스템의 흥미로운 확장은 그룹 추천시스템이다.
    - 이러한 경우 추천 시스템은 단일 사용자가 아닌 사용자 그룹에 특정 활동을 추천하도록 조정된다.
    - 예를 들어 단체로 영화나 텔레비전을 보는 것, 피트니스 센터에서 음악을 선택하는 것, 또는 단체 관광객들에게 추천하는 여행이 포함될 수 있다.
    - PolyLens와 같은 초기 시스템은 그룹 추천을 만들기 위해 개별 사용자의 선호도를 종합한 모델을 설계했다.
    - 그러나 수년에 걸친 합의로 **사용자의 선호도 합보다 다양한 사용자간의 상호 작용을 바탕으로 추천을 하는 추천 시스템 설계로 발전**했다.

> - 단순 평균화 전략은 그룹이 이질적이고 다양한 취향을 가진 사용자를 포함할 때 잘 작동하지 않는다.
    - 사용자가 감정 전염이나 순응 등 사회심리학에서 비롯된 현상을 바탕으로 서로의 기호에 영향을 미치는 경우가 많기 때문이다.
    
# 1.5.4 Multi-Criteria Recommender Systems

> - 다중 기준 시스템에서 등급은 단일 사용자에 의해 다른 기준에 기초하여 지정될 수 있다.
    - 예를 들어, 사용자는 줄거리, 음악, 특수 효과 등을 기준으로 영화를 평가할 수 있다.
    - 그러한 기법은 종종 항목에 대한 사용자의 만족을 다양한 기준에 해당하는 등급의 벡터로 모델링하여 권고사항을 제공한다.

> - 다중 기준 추천자 시스템에서는 기존의 추천자 시스템과 연계하여 전체 등급만 사용함으로써 잘못된 결과를 얻을 수 있는 경우가 많다.
    - 예를 들어, 두 사용자가 한 영화에 대한 전체 등급은 동일하지만 내용과 음악에 대한 구성 요소 등급은 매우 다르다면, 유사성 기반 협업 필터링 알고리즘의 관점에서 두 사용자를 유사하다고 간주해서는 안 된다.
- 일부 다중 기준 시스템에서 사용자는 전체 등급을 지정하지 않을 수 있다.
    - 이러한 경우에는 여러 기준을 기반으로 다양한 사용자에게 순위가 지정된 항목 목록을 제시해야하므로 문제가 더욱 어려워진다.
    
# 1.5.5 Active Learning in Recommender Systems

> - 추천 시스템의 주요 과제는 **강력한 예측을 하기 위해 충분한 등급을 획득하는 것**이다.
- 등급 행렬의 희소성은 추천 시스템의 효과적인 기능에 중요한 장애가 되고 있다.
    - 충분한 등급을 획득하면 희소성 문제를 줄일 수 있다.
- 다양한 실제 추천시스템은 사용자가 시스템을 채우기 위해 등급을 입력하도록 권장하는 메커니즘을 가지고 있다.
    - 예를 들어, 사용자에게는 특정 항목의 등급을 매기는 인센티브가 제공될 수 있다.
    - 일반적으로 취득과정의 비용이 높기 때문에 단일 사용자로부터 너무 많은 등급을 획득하기 어려운 경우가 많다.
    - 따라서 특정 사용자가 평가할 항목을 신중하게 선택해야 한다.
- 분류와 같은 다른 문제 영역에서는 접하지 않는 추천시스템에는 흥미로운 절충이 많이 있다.

# 1.5.6 Privacy in Recommender Systems

> - 추천시스템은 암묵적이거나 명시적일 수 있는 사용자들의 피드백을 기반으로 한다.
- 이 피드백은 **사용자의 `관심사`에 대한 중요한 정보를 포함**하고 있으며, 그들의 정치적 의견, 성적 지향 및 개인적 선호도에 대한 정보가 드러날 수 있다.
- **많은 경우, 그러한 정보는 매우 민감할 수 있으며, 이로 인해 `프라이버시` 문제가 된다.**
    - 이러한 개인 정보 보호 문제는 추천 알고리즘의 발전에 필요한 데이터의 공개를 방해한다는 점에서 중요하다.
- 실제 데이터의 가용성은 알고리즘 발전에 매우 중요하다.
    - 예를 들어, 추천 시스템 커뮤니티에 대한 Netflix Prize 데이터 세트의 기여는 많은 최신 알고리즘의 개발에 동기를 부여한 것으로 인정 될 수 있다는 점에서 매우 중요하다.
- 최근 몇 년 동안, 프라이버시의 주제는 다양한 데이터 마이닝 문제의 맥락에서 탐구되어 왔다.
    - 추천 도메인도 예외는 아니며 수많은 개인 정보 보호 알고리즘이 개발되었다.
    
# 1.5.7 Application Domains

> - 추천시스템은 소매, 음악, 콘텐츠, 웹 검색, 쿼리 및 컴퓨터 광고와 같은 수많은 응용 분야에서 사용된다.
    - 이러한 도메인 중 일부는 추천 시스템을 조정하기위한 특수한 방법이 필요하다.
    - 특히 13장에서는 뉴스 추천, 컴퓨터 광고, 상호 추천시스템에 해당하는 3개의 특정 영역을 연구한다.
    - 이 모든 애플리케이션 도메인은 본질적으로 **`웹` 중심**이다.
- 추천시스템의 중요한 측면은 장기적인 사용자 관심사를 추적하고 식별하기 위해 강력한 사용자 식별 메커니즘의 존재를 가정하는 것이다.
    - 많은 웹 도메인에서 강력한 사용자 식별을 위한 메커니즘을 사용할 수 없을 수 있다.
    - 이 경우 추천 기술의 직접사용이 불가능할 수 있다.
