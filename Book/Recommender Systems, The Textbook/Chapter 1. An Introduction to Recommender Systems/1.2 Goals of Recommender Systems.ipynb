추천 시스템의 목표를 논의하기 전에 추천 문제를 공식화 할 수있는 두 방법을 소개한다. 두 가지 기본 모델은 다음과 같다.

#### 1) Prediction version of problem
- 첫 번째 방법은 **user-item 조합에 대한 평가 값을 예측하는 것**이다. 


- item에 대한 user 선호도를 나타내는 train 데이터가 있다고 가정한다. m명의 사용자와 n개의 item의 경우, m × n의 불완전한 행렬이 되며, 명시된(또는 관측된) 값이 train에 사용된다. train된 모델을 사용해 결측값(또는 관찰되지 않은 값)을 예측할 수 있다. 


- 이 문제는 우리가 불완전하게 명시된 값의 행렬을 가지고 있고, 나머지 값은 학습된 알고리즘에 의해 예측해야 되기 때문에 **`matrix completion problem`**라고도 알려져 있다.

#### 2) Ranking version of problem
- 실제로 사용자에게 추천하기 위해 특정 항목에 대한 사용자의 평가를 예측할 필요가 없다. 대신, 판매자는 **특정 사용자에게 최상위 항목을 추천**하거나 특정 항목을 대상으로 할 몇몇 사용자에게만 추천을 할 수 있다. 


- 상위 k개의 item을 결정하는 것은 상위 k명의 사용자를 결정하는 것보다 일반적이지만, 두 경우 모두 방법은 정확히 유사하다. 이 책에서는 상위 k개의 item을 결정하는 것만을 논의할 것이다. 왜냐하면 그것이 더 일반적이기 때문이다. 


- 이 문제를 **`top-k recommendation problem`**라고도 한다.

첫 번째 경우가 더 일반적이다. 두 번째 경우에 대한 솔루션은 다양한 user-item 조합에 대한 첫 번째 문제를 풀고 예측 순위를 매김으로써 도출 될 수 있기 때문이다. 그러나 대부분의 경우 **`Ranking version of problem`** 을 직접 해결하는 방법을 설계하는 것이 더 쉽고 자연스럽다. 이 방법은 13장에서 논의될 것이다.

**제품 판매를 늘리는 것이 추천 시스템의 주요 목표**이다. 결국 **추천시스템은 판매자가 수익을 높이기 위해 활용**된다. 추천시스템은 적절한 item을 사용자에게 추천함으로써 사용자의 관심을 끌게한다. 이렇게 하면 판매자의 판매량과 수익이 증가합니다. 추천 시스템의 주요 목표는 판매자의 수익을 늘리는 것이지만, 이는 종종 첫 눈에 보이는 것보다 덜 명백하게 달성된다. 수익 증대라는 광범위한 business-centric 목표를 달성하기 위해 추천 시스템의 일반적인 운영 및 기술 목표는 다음과 같다.

#### 1) Relevance(관련성)
- **추천 시스템의 가장 분명한 운영 목표는 현재 사용자와 관련된 item을 추천하는 것**이다. 사용자는 자신이 흥미롭다고 생각하는 항목을 소비할 가능성이 더 높다. 


- 추천 시스템의 주요 운영 목표는 'relevance'이지만 이것만 고려하기는 충분하지 않다. 따라서, 아래의 몇 가지 이차 목표를 논의한다. 이 목표는 'relevance' 만큼 중요하지는 않지만, 그럼에도 불구하고 상당한 영향을 미칠 만큼 중요합니다.

#### 2) Novelty(참신한)
- **추천 시스템의 권장되는 item이 이전에 사용자가 보지 못한 항목일 때 매우 유용**하다. 예를 들어, 선호하는 장르의 인기 영화는 사용자에게 대부분 새로운 것이 아니다. 인기 품목의 반복적인 추천은 판매 다양성의 감소로 이어질 수도 있습니다.

#### 3) Serendipity(우연성)
- **serendipity(우연)**이라는 개념으로, **추천 항목이 다소 예상치 못한 것**이다. 'Serendipity' 는 사용자들이 이전에 몰랐던 것이 아니라, 사용자들에게 정말로 놀라운 추천이라는 점에서 'Novelty' 와는 다르다. 특정 사용자가 특정 유형의 itme만 소비하는 경우가 종종있을 수 있지만, 다른 유형의 item에 대한 잠재적인 관심이 존재해 사용자가 스스로 놀라게 할 수도 있다. 'Novelty' 와는 달리, 'Serendipity' 방법은 이러한 추천 사항을 발견하는 데 초점을 맞춘다.


- 예를 들어, 만약 새로운 인도 레스토랑이 이웃에 문을 연다면, 보통 인도 음식을 먹는 사용자에게 그 식당을 추천하는 것은 참신하지만 우연적인 것은 아니다. 반면에, 같은 사용자가 에티오피아 음식을 추천받았을 때, 그런 음식이 사용자에게 어필될 수 있다는 것을 사용자도 모를 때, 그 추천은 우연한 것이다.


- **Serendipity는 판매 다양성을 높이거나 사용자에 대한 새로운 관심 트렌드를 시작하는 유익한 부작용을 가지고 있다.**  'Serendipity'가 증가하면 새로운 관심 영역을 발견할 수 있기 때문에 종종 판매자에게 장기적이고 전략적인 이익을 가져다 준다. 반면, 우연히 추천하는 알고리즘은 **종종 관련 없는 item을 추천하는 경향이 있을 수 있다. 많은 경우, 우연한 방법의 장기적이고 전략적인 이점이 이러한 단기적인 단점을 능가한다.**

#### 4) Increasing recommendation diversity(다양성)
- 추천 시스템은 일반적으로 Top-k 항목을 제안한다. 이러한 **추천 항목이 모두 매우 비슷하면 사용자가 이러한 항목을 좋아하지 않을 위험이 증가**하게 된다. 반면 추천 목록에 서로 다른 유형의 항목이 포함되어 있으면 사용자가 이러한 항목 중 하나 이상을 좋아할 가능성이 더 높다. 


- **`다양성`은 사용자가 유사한 아이템을 반복해서 추천해도 지루하지 않도록 하는 장점이 있다.**

이러한 구체적인 목표 외에도, 사용자 및 판매자의 관점에서 추천 프로세스에 의해 다수의 소프트 목표도 충족된다. 사용자의 관점에서 추천시스템은 웹 사이트에 대한 전반적인 사용자 만족도를 향상시키는 데 도움이 될 수 있다. 예를 들어, 'Amazon.com' 에서 추천을 반복적으로 수신하는 사용자는 이 경험에 더 만족하고 사이트를 다시 사용할 가능성이 더 높다. 이렇게 하면 사용자 로얄티가 향상되고 매출이 더욱 늘어날 수 있게 된다. 판매자 측에서 추천 프로세스는 사용자의 요구 사항에 대한 통찰력을 제공하고 사용자 경험을 더욱 맞춤화하는 데 도움이 될 수 있다. 마지막으로, 사용자에게 특정 항목이 추천되는 이유에 대한 설명을 제공하는 것이 유용한 경우가 많다. 예를 들어 넷플릭스의 경우 이전에 본 영화와 함께 추천이 제공된다. 나중에 알게 되겠지만, 일부 추천 알고리즘은 다른 알고리즘보다 설명을 제공하는 데 더 적합하다는 것을 알 수 있다.

추천시스템의 추천하는 제품의 종류는 매우 다양하다. 페이스북과 같은 일부 추천시스템은 item을 직접 추천하지 않는다. 오히려 그들은 유용성과 광고 수익을 증가시켜 사이트에 간접적인 이익을 주는 소셜 연결을 추천할 수 있다. 이러한 목표의 특성을 이해하기 위해 과거 및 현재 추천 시스템의 몇 가지 인기있는 예를 논의 할 것이다. 이러한 예는 연구 프로토타입으로 구축되었거나 오늘날 다양한 문제 상황에서 상용 시스템으로 이용할 수 있는 다양한 추천시스템이다.

### GroupLens Recommender System

> GroupLens는 선구적인 추천 시스템으로 Usenet 뉴스 추천을 위한 연구 프로토 타입으로 구축되었다. **Usenet 독자로부터 rating을 수집해 다른 독자들이 기사를 읽기 전에 좋아할지 말지를 예측하는 시스템**이다. **가장 초기 자동화된 협업 필터링 알고리즘 중 일부는 GroupLens에서 개발되었다.** 이 그룹이 개발한 일반적인 아이디어로 책과 영화와 같은 다른 제품 설정으로도 확장되었다. 해당 추천 시스템을 각각 BookLens 및 MovieLens라고 불렸다. 협업 필터링 연구에 대한 선구적인 기여 외에도, GroupLens 연구 팀은 데이터 세트를 벤치마킹하기 쉽지 않았던 이 분야 초기 동안 여러 데이터 세트를 발표한 점이 눈에 띄었다.

### Amazon.com Recommender System
 
> Amazon.com은 특히 커머스 환경에서 추천 시스템의 선구자 중 하나였다. 초기에는 이 기술의 유용성을 실현할 수 있는 몇 안 되는 회사 중 하나였다. 원래 도서 전자 소매업자로 설립된 이 회사는 사실상 모든 형태의 상품으로 확대되었다. 결과적으로, 아마존닷컴은 현재 책, CD, 소프트웨어, 전자제품 등과 같은 거의 모든 범주의 제품을 판매하고 있다. **Amazon.com의 추천은 명시적으로 제공되는 rating, 구매 행동 및 검색 행동에 기초하여 제공된다.** Amazon.com의 등급은 5점 척도로 지정되며, 최저 등급은 1성, 최고 등급은 5성이다. 아마존이 지원하는 account authentication mechanism으로 사용자가 로그인하면 고객별 구매 및 로그 데이터를 손쉽게 수집할 수 있다. **사용자들이 계정에 로그인할 때마다 사이트의 메인 웹페이지에서 추천 항목이 제공**된다. 추천에 대한 설명이 제공되는 경우가 많다.

> **사용자의 구매, 검색 로그는 사용자가 지정한 명시적(`explicit`) 등급과는 반대로 암묵적(`implicit`) 등급의 한 종류로 볼 수 있다.** 많은 커머스 시스템은 명시적 피드백과 암묵적 피드백에 기초하여 추천을 제공하는 유연성을 허용한다. 실제로 여러 모델은 추천 과정에서 명시적 피드백과 암묵적 피드백을 공동으로 설명하도록 설계되었다.

### Netflix Movie Recommender System

> Netflix는 영화 및 TV 쇼의 우편 주문 디지털 비디오 디스크(DVD) 대여 회사로 설립되었으며 결국 스트리밍 전송으로 확장되었다. 현재 Netflix의 주된 사업은 영화와 텔레비전 쇼의 스트리밍을 구독제로 제공하는 사업이다. Netflix는 사용자에게 영화와 텔레비전 쇼의 등급을 5점 만점으로 매길 수 있는 기능을 제공한다. 또한 다양한 항목을 시청하는 사용자의 행동도 저장이 된다. 이러한 **rating과 행동을 기반으로 추천**을 한다. Netflix는 추천 아이템에 대한 설명을 훌륭히 해내고 있다. 사용자가 시청한 특정 항목을 바탕으로 추천 예시를 명시적으로 제공한다. 이러한 정보는 사용자에게 특정 영화를 볼 것인지 여부를 결정하기 위한 추가 정보를 제공한다. **의미 있는 설명을 제시하는 것은 사용자에게 왜 특정 영화가 흥미롭다고 생각하는지에 대한 이해를 제공하기 위해 중요**하다. 이 접근방식은 사용자가 추천 항목에 따라 행동할 가능성이 더 높아지고 사용자 경험을 진정으로 향상시켜준다. 이러한 유형의 흥미로운 접근방식은 고객 충성도와 유지력 향상에 도움이 될 수 있다.

> Netflix는 Netflix Prize 콘테스트로 연구 커뮤니티에 크게 기여했습니다. 이 대회는 참가자들이 기여한 다양한 협업 필터링 알고리즘 간의 경쟁을 위한 포럼을 제공하기 위해 고안되었습니다. 넷플릭스 영화 rating의 데이터 세트가 공개되었고, 과제는 특정 사용자의 item rating을 예측하는 것이었다. 이를 위해 넷플릭스는 train data와 qualifying data를 모두 제공했다. 사용자는 훈련 데이터에 기반한 모델에 기초해 qualifying data의 rating을 예측해야 했다.

### Google News Personalization System

> **Google News Personalization System은 클릭 이력을 바탕으로 사용자에게 뉴스를 추천한다.** 클릭은 Gmail 계정에 의해 활성화된 식별 메커니즘을 기반으로 특정 사용자와 연관된다. 이 경우 뉴스 기사는 item으로 취급된다. 사용자가 뉴스 기사를 클릭하는 행위는 해당 기사에 대한 긍정적인 평가로 볼 수 있다. 이러한 rating은 **사용자가 항목에 대한 호감을 표시할 수 있는 메커니즘이 존재하지만, 싫어하는 것을 나타낼 수 있는 메커니즘은 존재하지 않는 `unary rating`**으로 볼 수 있다. 더욱이, rating은 사용자에 의해 명시적으로 지정되기 보다는 사용자 행동으로부터 유추되기 때문에 암묵적(implicit)이다. 수집 된 rating을 협업 추천 알고리즘에 적용해 특정 사용자를 위한 개인화된 기사에 대한 추론을 할 수 있게 되었다.

### Facebook Friend Recommendations

> **소셜네트워크서비스(SNS)는 사이트 내 소셜 커넥션 수를 늘리기 위해 잠재적인 친구를 이용자에게 추천**하는 경우가 많다. 페이스북은 소셜 네트워크 웹사이트의 한 예다. 이런 종류의 추천은 제품 추천과는 약간 다른 목표를 가지고 있다. 상품 추천은 상품 판매를 촉진하여 직접적으로 가맹점의 이익을 증대시키는 반면, 소셜 커넥션의 수가 증가하면 소셜 네트워크 이용자의 경험이 향상된다. 이는 결국 소셜 네트워크의 성장을 촉진한다. **소셜 네트워크는 광고 수익을 늘리기 위해 네트워크의 성장에 크게 의존**하고 있다. 그러므로 잠재적 친구(또는 링크)의 추천은 네트워크의 더 나은 성장과 연결을 가능하게 한다. 이 문제를 소셜 네트워크 분석 분야에서 **link prediction**으로 알려져 있다.그러한 추천의 형태는 rating 데이터보다는 **구조적 관계에 기초**한다. 그러므로 기본 알고리즘의 성격이 전혀 다르다. 링크 추천 문제는 10장에서 얘기할 것이다.
