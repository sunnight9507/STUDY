전자 및 비즈니스 거래를 위한 매체로 **Web의 중요성이 증대됨에 따라 추천시스템 기술 개발의 원동력이 되었다.** 이와 관련하여 중요한 촉매제는 웹이 사용자에게 호불호에 대한 **`피드백`**을 쉽게 제공받을 수 있도록 하는 것이다.

예를 들어 넷플릭스와 같은 콘텐츠 공급자의 시나리오를 생각해보자.

이 경우 사용자는 마우스를 클릭하기만 하면 쉽게 피드백을 제공할 수 있다. 피드백을 제공하는 일반적인 방법론은 사용자가 다양한 항목에 대한 **호불호를 지정하는 평가 시스템(explicit data)**(예: 별 5개 rating 시스템)에서 숫자를 선택하는 형태이다.

다른 형태의 피드백은 **implicit data**로 Web-centric paradigm에서 수집하기 쉽다.

예를 들어, 사용자가 항목을 구입하거나 검색하는 간단한 행위를 해당 항목으로 간주될 수 있다. 이러한 형태의 피드백은 Amazon.com과 같은 온라인 상점에서 일반적으로 사용되며, 이런 종류의 데이터 수집은 고객이 필요로 하는 작업이라는 점에서는 매우 용이하다.

**추천시스템의 기본 개념은 이러한 다양한 데이터를 활용하여 `고객의 관심`을 `추론`하는 것이다.**

추천이 제공되는 주체를 'user', 추천되는 제품을 'item'이라고 한다.

따라서, **추천 분석**은 종종 **사용자와 아이템의 이전의 상호작용을 기반**으로 한다. 과거의 관심사와 성향은 종종 미래의 선택을 나타내는 좋은 지표이기 때문이다. 예외로 **`knowledge-based recommender systems`**이며, 사용자의 과거 이력이 아닌 **사용자 지정 요구사항**에 따라 권장사항을 제안하는 경우가 있다.

그렇다면 추천 알고리즘의 기초가 되는 기본 원칙은 무엇일까?

> **추천 시스템의 기본 원칙은 사용자와 아이템 사이에 상당한 종속성이 존재한다는 것**

예를 들어, 역사 다큐멘터리에 관심이 있는 사용자일 경우 액션 영화보다는 다른 역사 다큐멘터리나 교육 프로그램에 더 관심을 가질 가능성이 높다. 

많은 경우 다양한 범주가 중요한 상관관계를 나타내므로 보다 정확한 추천을 할 수 있다. 또는 종속성이 범주가 아닌 개별 항목의 세분화로 나타날 수 있다. 이러한 종속성은 rating matrix에서 data-driven 방식으로 학습할 수 있으며 결과 모델은 대상 사용자를 예측하는 데 사용된다. 사용자가 이용할 수 있는 등급 항목의 수가 많을수록 사용자의 향후 동작에 대한 예측을 쉽게 할 수 있다. 다양한 모델을 사용해 이 작업을 수행할 수 있다.

예를 들어, 공동으로 구매한 아이템 또는 rating을 활용해 유사한 제품에 관심이 있는 사용자 간의 군집을 만들 수 있다. 이러한 군집의 관심사와 행동을 활용하여 군집의 개별 사용자에게 추천할 수 있다. 

앞서 언급 한 설명은 neighborhood models로 매우 간단한 추천 알고리즘이고 **Collaborative filtering(협업 필터링)**이라고 하는 더 광범위한 모델에 속한다. **"협업 필터링"** 이라는 용어는 여러 사용자의 rating을 collaborative한 방법으로 비어있는 rating 값을 예측하는 것을 의미한다.

실제로 추천 시스템은 다양한 데이터 유형으로 복잡하고 많다. 예를 들어 콘텐츠 기반 추천 시스템에서 콘텐츠는 추천 프로세스에서 주요 역할을 하며, 사용자 rating과 항목의 속성을 활용해 예측한다. 기본 아이디어는 **사용자가 과거에 평가를 했거나 접근한 항목의 속성을 기반**으로 **사용자의 관심사를 모델링** 할 수 있다는 것이다. 다른 프레임 워크로는 **사용자가 대화식으로 자신의 관심사를 선택하고 도메인 지식과 결합되어 추천**을 하는 **knowledge-based systems**이 있다. 발전되고 있는 추천 모델에서는 시간 정보, 외부 정보, 위치 정보, 사회 정보 또는 네트워크 정보와 같은 상황별 데이터를 사용할 수 있다.

이 책은 협업, 콘텐츠 기반, 지식 기반 추천 시스템을 포함한 모든 유형의 기본 시스템에 대해 알아갈 것이다. 또한, 서로 다른 도메인에서 추천시스템의 기본 및 향상된 모델에 관해 설명할 것이다. 추천 시스템의 신뢰할 수 있는 모델들의 구조에 대해 다양한 측면으로 연구할 것이다. 또한 추천 시스템에 대한 다양한 평가 및 hybrid 모델을 알아갈 것이다. 이 장의 목표는 추천 시스템 분야의 광범위한 작업에 대한 개요를 제공하고 다양한 주제를 개별 장과 관련시키는 것이다.
