```python
# 74점
def solution1(N, stages):
    stage_num = []
    failure_rate = {}
    
    for i in range(0,N):
        stage_num.append(0)
        failure_rate[i] = 1
    
    for i in stages:
        for j in range(0,i-1):
            stage_num[j] += 1
        
    init_num = len(stages)
    
    cnt = 1
    
    for i in stage_num:
        failure_rate[cnt] = 1 - i/init_num
        cnt += 1
        init_num = i
        
    #print(failure_rate)
    
    result_rate = sorted(failure_rate.items(), key = (lambda x:x[1]), reverse = True)
    
    #print(result_rate)
    
    answer = []
    
    for i in result_rate:
        if i[0] == 0:
            continue
        answer.append(i[0])
    
    
    return answer
```

```python
# 70.4
def solution(N, stages):    
    stages.sort(reverse = True)
    
    print(stages)
    
    fail_rate = {}
    
    len_stages = len(stages)
    
    start_level = N
    
    index = 0
    
    same_level = 0
    
    while(start_level > 0):
        if index == len(stages):
            fail_rate[start_level] = same_level / len_stages
            same_level = 0
            start_level -= 1
            continue
        if stages[index] > start_level:
            index += 1
        elif stages[index] == start_level:
            index += 1
            same_level += 1
        else:
            fail_rate[start_level] = same_level / index
            same_level = 0
            start_level -= 1
    
    fail_rate = dict(sorted(fail_rate.items(), key = (lambda x:x[0])))
    
    result_rate = sorted(fail_rate.items(), key = (lambda x:x[1]), reverse = True)
    
    answer = []
    
    for i in result_rate:
        if i[0] == 0:
            continue
        answer.append(i[0])
    
    return answer
```

```python
# 70.4
import functools

def compare(ob1, ob2):
    return ob2[1] - ob1[1]

def solution(N, stages):    
    
    level_num = []
    
    for i in range(0,N+2):
        level_num.append(0)
        
    for i in stages:
        level_num[i] += 1
    
    fail_rate = []
    
    total_num = len(stages)
    
    index = 1
    
    for i in level_num[1:N+1]:
        fail_rate.append([index, i / total_num])
        total_num -= i
        index += 1
    
    #print(fail_rate)

    answer = [x[0] for x in sorted(fail_rate, key=functools.cmp_to_key(compare))]
    
    return answer
```

```python
from collections import Counter as mset
import functools

def compare(ob1, ob2):
    return ob2[1] - ob1[1]

def solution(N, stages):
    answer = []
    failureRateLst = []
    stageSet = mset(stages)    
    for stage in range(0, N):
        tryer = 0
        finisher = 0
        for key in stageSet.keys():
            print(key)
            if key-1 == stage:
                tryer += stageSet[key]
            elif key-1 > stage:
                finisher += stageSet[key]
        if (tryer+finisher) != 0:
            failureRateLst.append([stage+1, tryer/(tryer+finisher)])
        else: failureRateLst.append([stage+1, 0])
            
    print(failureRateLst)
    answer = [x[0] for x in sorted(failureRateLst, key=functools.cmp_to_key(compare))]
    
    return answer
```
