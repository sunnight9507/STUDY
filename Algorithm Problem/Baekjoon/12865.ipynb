### [평범한 배낭](https://www.acmicpc.net/problem/12865)

```python
N, K = map(int, input().split())
result = [0 for _ in range(K+1)]

for _ in range(N):
#     print(result)
    w, v = map(int, input().split())
    
    if w > K: continue
    tmp_lst = [0 for _ in range(K+1)]
    for _ in range(w, K+1):
        if result[_] < v + result[_ - w]:
            tmp_lst[_] = v + result[_ - w]
        else: tmp_lst[_] = result[_]
    result[w:] = tmp_lst[w:]

# print(result)
print(result[-1])
```
